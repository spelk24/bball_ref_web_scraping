```python
import pandas as pd
import bs4
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
#Can use this with advanced stats or regular per game stats
my_url = 'https://www.basketball-reference.com/leagues/NBA_2019_advanced.html'
```

```python
#Download HTML page
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
```

```python
#HTML parsing
page_soup = soup(page_html,"html.parser")
```

```python
#Table Container
table = page_soup.find("div",{"class":"table_outer_container"})
```

```python
#Scrape Column Names
headers = table.div.table.tr.findAll("th")
header_list = [th.text for th in headers]
```

```python
#Each row in the table
player_rows = table.div.table.tbody.findAll("tr")
len(player_rows)
```

```python
#Stat Scraper
new = pd.DataFrame(columns = header_list)
new = new.drop('Rk',axis =1)



for row in player_rows:
    comma_sep_stats = []
    try:
        player_info = row.findAll("td")
        for stat in player_info:
            comma_sep_stats.append(stat.text)
        new.loc[len(new)] = comma_sep_stats
    except:
        pass

    
    
len(new)
```

```python
df = new.copy()
df.drop('\xa0', axis = 1, inplace= True)
cols = df.columns.tolist()[4:]
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df2 = df.sort_values(by= 'MP', ascending = False).reset_index(drop = True)
```
